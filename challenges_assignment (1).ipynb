# Challenge 4: Print out all the indices of the non-zero elements in this array: [6,0,9,0,0,5,0]

```python
# write your code here
import numpy as np
arr=np.array([6,0,9,0,0,5,0])
non_zero_indices=np.nonzero(arr)
print(non_zero_indices[0])
```
